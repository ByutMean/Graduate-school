**16번. 적게 등장한 나이 중 최연소**

```python
grouped = data.groupby(data['age']).size() # 나이들의 등장 횟수 추출 방법1
```

```python
grouped = data['age'].value_counts() # 나이들의 등장 횟수 추출 방법2
```

```python
grouped
```

```python
grouped.values
```

```python
grouped.index
```

```python
result=[]
for i,j in zip(grouped.values, grouped.index):
  if i==grouped.values.min():
    result.append(j)
```

```python
result # 적게등장한 나이들
```

```python
min(result) #중 최연소
```
