```python
from konlpy.tag import Kkma, Twitter, Hannanum
from pprint import pprint
```

```python
doc = open('C:\\Users\\AAA\\Downloads\\aaaa.txt', 'r', encoding='utf-8').read()

print(doc)
```

```python
pos_tagger = Hannanum()

pos = ['/'.join(t) for t in pos_tagger.pos(doc)] 

pos_nouns = ['/'.join(t) for t in pos_tagger.pos(doc) if t[1]=='N']

pprint(pos[0:20])

pprint(pos_nouns[0:20])
```

```python
text = nltk.Text(pos_nouns) 
text
```

```python
text.tokens
```

```python
len(set(text.tokens))
```

```python
text.vocab().most_common(20)
```

```python
print("문서에 나온 토큰 수: ", len(text.tokens)) 
print("문서에 나온 토큰 종류 수:  ", len(set(text.tokens)))
pprint(text.vocab().most_common(20))
```
