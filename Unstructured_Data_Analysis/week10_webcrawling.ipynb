```python
import requests
from selenium import webdriver
import time
import csv
from bs4 import BeautifulSoup
```

```python
browser = webdriver.Chrome()
def get_reple(page):
    url = "https://movie.naver.com/movie/bi/mi/pointWriteFormList.nhn?code=190010&type=after&onlyActualPointYn=Y&onlySpoilerPointYn=N&order=sympathyScore&page={}"
    browser.get(url.format(page))
    response = requests.get(url.format(page))
    soup = BeautifulSoup(response.text, 'html.parser')
    for page in range(1,5):

        reviews=[]
        sample2= browser.find_elements_by_class_name('btn_more')
        for kk in range(len(sample2)):
            sample2[kk].click()
            kk=kk+1

       
        sample= browser.find_elements_by_class_name('_unfold_ment')

        for j in range(len(sample)):
            sample[j].click()
            j=j+1   

        rev0=browser.find_element_by_css_selector('#_filtered_ment_0')
        review0=rev0.text
        rev1=browser.find_element_by_css_selector('#_filtered_ment_1')
        review1=rev1.text
        rev2=browser.find_element_by_css_selector('#_filtered_ment_2')
        review2=rev2.text
        rev3=browser.find_element_by_css_selector('#_filtered_ment_3')
        review3=rev3.text
        rev4=browser.find_element_by_css_selector('#_filtered_ment_4')
        review4=rev4.text
        rev5=browser.find_element_by_css_selector('#_filtered_ment_5')
        review5=rev5.text
        rev6=browser.find_element_by_css_selector('#_filtered_ment_6')
        review6=rev6.text
        rev7=browser.find_element_by_css_selector('#_filtered_ment_7')
        review7=rev7.text
        rev8=browser.find_element_by_css_selector('#_filtered_ment_8')
        review8=rev8.text
        rev9=browser.find_element_by_css_selector('#_filtered_ment_9')
        review9=rev9.text        

        reviews.append(review0)
        reviews.append(review1)
        reviews.append(review2)
        reviews.append(review3)
        reviews.append(review4)
        reviews.append(review5)
        reviews.append(review6)
        reviews.append(review7)
        reviews.append(review8)
        reviews.append(review9)

        return reviews
    
    browser.close()
```

```python
import time
text = []
for i in range(1,5):
    time.sleep(1)
    print(i,end='\r')
    reviews = get_reple(i)
    text += reviews
```

```python
import pandas as pd

df = pd.DataFrame([text]).T
df.columns = ['text']
df
```
