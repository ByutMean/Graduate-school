```python
from konlpy.tag import Twitter
import pandas as pd
new_corpus = pd.read_csv('C:\\Users\\AAA\\Downloads\\dataset11.csv', encoding=' cp949') #특허 데이터 활용
```

```python
new_corpus
```

```python
real = new_corpus['text']
```

```python
pos_tagger=Twitter()
import re
total_news = []
stopword = ['것','수','저']

for news in real[:1000]:
    pos_news = re.sub('.* 기자 = ', '', news)
    pos_news = ['/'.join(t[:-1]) for t in pos_tagger.pos(pos_news) if ((t[1]=='Noun') & (t[0] not in stopword))]
    total_news.append(' '.join(pos_news))
```

```python
from wordcloud import WordCloud
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
kor_vectorizer = CountVectorizer(min_df=10)
kor_bow = kor_vectorizer.fit_transform(total_news)
```

```python
transformer = TfidfTransformer()
tfidf = transformer.fit_transform(kor_bow.toarray())
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=5, n_init=50, verbose=1)
```

```python
kmeans.fit(tfidf)
```

```python
kmeans.labels_
```

```python
import numpy as np
clusters = []
for i in range(0, kmeans.n_clusters):
    clusters.append(np.where(kmeans.labels_ == i))
```

```python
import gensim


#아래 명령어 실행하시고
def hash32(value):
     return hash(value) & 0xffffffff
    
doc_list = []
for doc in real[:1000]:
    tokens = [t for t in pos_tagger.nouns(doc) if len(t)>1]
    doc_list.append(tokens)
```

```python
import time
min_count = 1 #2 번 미만으로 나온 단어는 무시하겠다
hidden_size = 50 #corpus 가 적을때는 50, 30 정도만 해도 충분하다, 많으면 100~200
workers = 1 #core 수
window= 3 #corpus가 작으면 window size가 적은게 좋고, 많으면 5~10으로 해도 됨
epoch = 30  #30~50정도

start = time.time()
model_total = gensim.models.Word2Vec(doc_list, window = window, min_count=min_count, size = hidden_size, workers = workers, iter = epoch, seed=1, hashfxn=hash32)

end = time.time()
print(end-start)
```

### classfication

```python
data= pd.read_csv(r'C:\Users\AAA\Downloads\real_tenet.csv',encoding='utf-8')
```

```python
data
```

```python
corpus=data['text']
```

```python
from konlpy.tag import Kkma
prr=[]
pos_tagger = Kkma()
for i in corpus:
    pos_r=pos_tagger.pos(i)
    c=[]
    for t in pos_r:
        if ((((t[1]=='NN'))or ((t[1]=='NNG'))or ((t[1]=='NNB'))or ((t[1]=='NNM'))or ((t[1]=='NNP'))or ((t[1]=='NP')))):
            if not (len(t[0])==1):
                c.append(t[0])
    prr.append(c)
```

```python
result=[]
for i in prr :
    prr_join=" ".join(i)
    result.append(prr_join)
```

```python
result
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
x = vectorizer.fit_transform(result)
```

```python
aspect_word = '스토리'
aspect_id = vectorizer.vocabulary_.get(aspect_word, -1)

print('%s = %d' % (aspect_word, aspect_id))
print('num of doc (%s) = %d' % (aspect_word, len(x[:,aspect_id].nonzero()[0])))
```

```python
print(x[:,aspect_id].shape)

from pprint import pprint
pprint(x[:,aspect_id].nonzero())
```

```python
pos_idx = set(x[:,aspect_id].nonzero()[0]) # in 함수는 list보다 set이 빠릅니다
y = [1 if i in pos_idx else -1 for i in range(x.shape[0])]

print('x shape = %s, len(y) = %d, num_pos = %d' % (str(x.shape), len(y), len(pos_idx)))
```

### LDA

```python
from gensim import corpora
dictionary = corpora.Dictionary(prr)
corpus = [dictionary.doc2bow(text) for text in prr]
print(corpus) 
```

```python
import gensim
NUM_TOPICS = 8 
ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics = NUM_TOPICS, id2word=dictionary, passes=15)
topics = ldamodel.print_topics(num_words=10)
for topic in topics:
    print(topic)
```
